```python
import pandas as pd
```

```python
df = pd.read_csv('features.csv')
```

```python
df.columns
```

```python
cols = ['gunning_fog', 'smog_index',
       'subjectivity', 'title_text_similarity', 'words_per_sentence',
       'adv_ratio_title', 'noun_ratio_title', 'words_per_title',
       'syllables_per_word', 'uppercase_per_title',
       'uppercase_ratio_title_text', 'adj_ratio_text', 'adv_ratio_text',
       'verb_ratio_text', 'noun_ratio_text', 'short_sentences_ratio',
       'long_sentences_ratio']
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

pca = PCA()
X = df[cols].values
X_scaled = StandardScaler().fit_transform(X)
X_pca = pca.fit(X_scaled)

explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)
explained_variance = pca.explained_variance_  # pentru Kaiser


plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.title('Varianta Cumulată Explicată de PCA')
plt.xlabel('Numărul Componentelor')
plt.ylabel('Varianta Cumulată')
plt.grid(True)
plt.show()

```
